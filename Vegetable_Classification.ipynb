```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
! unzip /content/vegetable-image-dataset.zip
```

```python
train_data_gen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
test_data_gen = ImageDataGenerator(rescale=1./255)
```

```python
seed=42
target_size=(224, 224)
batch_size= 32

train_data = train_data_gen.flow_from_directory(directory='/content/Vegetable Images/train', seed=42, target_size=target_size,batch_size=batch_size, subset="training" )
val_data = train_data_gen.flow_from_directory(directory='/content/Vegetable Images/train', subset='validation', seed=42, target_size=target_size,batch_size=batch_size)
test_data = test_data_gen.flow_from_directory(directory='/content/Vegetable Images/test', seed=42, target_size=target_size,batch_size=batch_size )
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten
```

```python
tf.random.set_seed(seed)

model1 = Sequential([
                     Conv2D(10, 3, input_shape=(224,224, 3), activation="relu"),
                     MaxPool2D(pool_size=2 ),
                     Conv2D(10, 3, activation="relu"),
                     MaxPool2D(pool_size=2 ),
                     MaxPool2D(pool_size=2 ),
                     Flatten(),
                     Dense(100, activation='relu'),
                     Dense(15, activation='softmax')
])


model1.compile(loss='categorical_crossentropy', optimizer="Adam", metrics=['accuracy'])

model1.fit(train_data,epochs=9, steps_per_epoch=len(train_data), validation_steps=len(val_data), validation_data=val_data )
```

```python
model1.evaluate(test_data)
```

```python
from tensorflow.keras.preprocessing import image

def convert_for_model(img_path,model, target_size=target_size):
    img = image.load_img(img_path, target_size=target_size)
    x = image.img_to_array(img)
    x = x.reshape((1,)+ x.shape)
    x = x/255
    pred = model.predict([x])
    res = pred.argmax()

    for key in classes:
        if classes[key] == res:
          return key
```
